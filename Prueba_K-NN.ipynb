###Separar las caractersticas (X) y la variable objetivo (y)



```python
X = data.drop(['Ingreso', 'Ingreso_encoded'], axis=1)  # Elimina las columnas 'Ingreso' y 'Ingreso_encoded'
y = data['Ingreso_encoded']  # La variable objetivo es 'Ingreso_encoded'
```

###EQUILIBRAR LA DATA

```python
# Crear un objeto RandomOverSampler
oversampler = RandomOverSampler(random_state=42)

# Aplicar el sobremuestreo a los datos
X_resampled, y_resampled = oversampler.fit_resample(X, y)
```

```python
# Crear un nuevo DataFrame con los datos equilibrados
balanced_data = pd.DataFrame(X_resampled, columns=X.columns)
balanced_data['Ingreso_encoded'] = y_resampled
# Verificar el equilibrio de las clases
print(balanced_data['Ingreso_encoded'].value_counts())
```

###TRATAMIENTO DE LA DATA

```python
# Identify categorical columns
categorical_cols = X_resampled.select_dtypes(include=['object']).columns

# Create a ColumnTransformer to handle categorical and numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X_resampled.select_dtypes(exclude=['object']).columns),
        ('cat', OneHotEncoder(), categorical_cols)
    ])

# Fit and transform the data
X = preprocessor.fit_transform(X_resampled)
```

### DIVIDIR LA DATA

```python
# prompt: separa la data en train y test

# Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y_resampled, test_size=0.2, random_state=2)
```

```python
# Aplicar SMOTE para balancear los datos separados en test y tarin

smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)
#X_test, y_test = smote.fit_resample(X_test, y_test)
```

```python
print('Numero de muestras en y_train:', np.bincount(y_train))
#print('Numero de muestras en y_test:', np.bincount(y_test))
```

```python
# Normalizar caractersticas
scaler = StandardScaler(with_mean=False)  # Set with_mean to False
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### MODELO DE K-NN


```python
# Crear modelo KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
from sklearn.decomposition import TruncatedSVD

# Reducir dimensionalidad a 2 componentes
# Alternatively, use TruncatedSVD which supports sparse matrices
svd = TruncatedSVD(n_components=2)
X_pca = svd.fit_transform(X_train_scaled)

# Crear un DataFrame con los datos reducidos y las etiquetas de cluster
df_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
df_pca['Ingreso_encoded'] = y

# Graficar los clusters en 2D
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Ingreso_encoded', data=df_pca, palette='viridis', s=80)
plt.title('Clusters en 2D (PCA)')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
# Evaluar modelo
from sklearn.metrics import accuracy_score, classification_report

# Predecir en el conjunto de prueba
y_pred_knn = knn.predict(X_test_scaled)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f'Accuracy KNN: {accuracy_knn:.2f}')
print(classification_report(y_test, y_pred_knn))
```

```python

# Predecir en el conjunto de entrenamiento
y_train_pred_knn = knn.predict(X_train_scaled)
train_accuracy_knn = accuracy_score(y_train, y_train_pred_knn)
print(f'Train Accuracy KNN: {train_accuracy_knn:.2f}')

# Predecir en el conjunto de prueba
y_test_pred_knn = knn.predict(X_test_scaled)
test_accuracy_knn = accuracy_score(y_test, y_test_pred_knn)
print(f'Test Accuracy KNN: {test_accuracy_knn:.2f}')
```

```python
from sklearn.metrics import confusion_matrix
# Calcular la matriz de confusin
cm = confusion_matrix(y_test, y_pred_knn)

# Graficar la matriz de confusin
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusin - KNN')
plt.xlabel('Predicciones')
plt.ylabel('Valores Reales')
plt.show()
```
